```python
import pandas as pd

# list all csv files only
csv_files = []

for item in range(0, 100):
    csv_files.append("files/random_"+str(item)+".csv")

csv_files
```

```python
df_csv_append = []
for file in csv_files:
    df = pd.read_csv(file)
    #df = df.head(2400)
    if(len(df)==1801):
        df = df.head(1800)
    if(len(df)==3601):
        df = df.head(3600)
    if(len(df)==5401):
        df = df.head(5400)
    df_csv_append.append(df)

```

```python
final_csv = pd.concat(df_csv_append, ignore_index=True)
final_csv.to_csv("files/dataset_random_trajectory.csv", index=False)
```
